```python
from threading import Thread
```

```python
#program for calling simple functions (independent)

def function1():
    for i in range(100):
        print("python")
        
def function2():
    for i in range(100):
        print("java")

function1()
function2()
```

```python
#program for calling simple functions (independent) ===> thread ===> parallel execution

def function1():
    for i in range(100):
        print("python")
        
def function2():
    for i in range(100):
        print("java")

function1()
function2()
```

```python
a = Thread(target = function1)
b = Thread(target = function2)
```

```python
a.start()
b.start()
print("main thread")
```

```python
#program for calling simple functions (independent) ===> thread ===> parallel execution

def function1():
    for i in range(100):
        print("python")
        
def function2():
    for i in range(100):
        print("java")
        
def function3():
    for i in range(100):
        print("groovy")

a = Thread(target = function1)
b = Thread(target = function2)
c = Thread(target = function3)


a.start()
b.start()
a.join() #waiting for completion of a thread ==> join function
b.join()
c.start()
print("main thread")
```

```python
#program for calling simple functions (independent) ===> thread ===> parallel execution #passing arguments

def function1(p):
    for i in range(100):
        print(p)
        
def function2(q):
    for i in range(100):
        print(q)
        
def function3(x):
    for i in range(100):
        print(x)

a = Thread(target = function1, args = ("c",))
b = Thread(target = function2, args = ("cpp",))
c = Thread(target = function3, args = ("js",))


a.start()
b.start()
c.start()
print("main thread")
```

#### polymorphism

```python
#function or method or conceptually something accepts more than one form

#inbuilt function len

print(len("python"))
print(len([1,2,3,5]))

#accepts both string and list input (accepts other data structure)
```

```python
#user defined function

def func1(x,y, z=0):
    return x + y + z

print(func1(3,4))
print(func1(3,4,5))

#Accepts both two params and three params
```

```python
#any ===> OR gate and all ===> AND gate ==> inbuilt functions in python

a = True
b = False
c = True

print(any((a,b,c)))
print(all((a,b,c)))
```

```python
#any and all
a = True
b = True
c = True

print(any((a,b,c)))
print(all((a,b,c)))
```

```python
#specialised sorts

x = [4,5,6,7,-1,2,3,9,1]

#default ascending order
print(sorted(x))

#reverse arguement for changing order
print(sorted(x, reverse = True)) #descending order
```

```python
#frozen set
#immutable iterable object ===> ordering, removes duplicates, immutable

x = [4,5,6,7,-1,2,3,9,1, 1]

z = frozenset(x)

print(z)
print(type(z))
```

```python
#shallow copy and deep copy ===> important interview question module copy

x = [4,5,6,7,-1,2,3,9,1]

#shallow copy
y = x

#deep copy
z = x.copy()

x[1] = 100

print(id(x))
print(x)
#shallow copy (will alter when parent changes) ===> pointing same storage
print(id(y))
print(y)
#deep copy (will not alter when parent changes) ===> pointing new storage
print(id(z))
print(z)
```

```python
#iterator (for loop basic structure)

x = [4,5,6,7,-1,2,3,9,1]

val = iter(x)

print(next(val))
print(next(val))
print(next(val))
```

```python
#Generator #iterator + return #==> return is the last line in function

def func1():
    yield "first"
    yield 2
    yield 23.0
    
for i in func1():
    print(i)
```

```python
#decorator
#one function available with exception or enhancement 

#decorator function
def supporting_func(func):
    def inner_func():
        print("line 1")
        func()
        print("line 2")
    return inner_func

#function that we cant alter #available in some other system or module 
def actual_func():
    print("Actual")
    
actual_func = supporting_func(actual_func)
actual_func()
```

```python
#__main__ super function
# if something comes with __function__ ==> super function

def func1(x,y):
    return x + y

print(__name__)

if __name__ == "__main__":
    print(func1(3,4))
```

```python
#dynamic args *

def func1(*args1):
    print(args1)
    for i in args1:
        print(i)
    
#with two args
print(func1(3,4))

#with three args
print(func1(3,4,5))

#with one args
print(func1(3))
```

```python
#dynamic args **

def func2(**args2):
    print(args2)
    for i in args2:
        print(i)
        print(args2[i])
    
print(func2(val1="hi", val2="hello"))
print(func2(val1="hi", val2="hello", val3="gggg", val4="hhh"))
```
